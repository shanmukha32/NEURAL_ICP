```python
from collections import Counter

#function for wordcount
def count_words_file(input_file, output_file):
# Counter to count word occurrences
  word_counter = Counter()

  # read the input file and process each line
  with open(input_file, 'r') as file:
    lines = file.readlines()
  for line in lines:  # Collect words from all lines
    words = line.strip().split()
  word_counter.update(words)

  # original text and the word count to the output file
  with open(output_file, 'w') as file:
    with open(input_file, 'r') as infile:
      file.write(infile.read())  # Write the original content

    #word count
    file.write("\nWord_Count:\n")
    for word, count in word_counter.items():
      file.write(f"{word}: {count}\n")


count_words_file('input.txt', 'output.txt')
```
